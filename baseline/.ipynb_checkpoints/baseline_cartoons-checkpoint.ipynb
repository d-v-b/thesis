```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.morphology as morph
from scipy.ndimage import correlate
```

```python
dsize = 3,64,64
bg = 1 + np.random.randn(*dsize) / 2
#bg = np.ones(dsize)
cellpos = [(20,20),(40,40)]
cell_bl = (0,0,0),(12,12,12)
cell_fluc = (1,2,3), (0,0,0)
rad = 8
cells_raw = np.zeros(bg.shape)
cells_bl = np.zeros(bg.shape)
for tind in range(cells_raw.shape[0]):
    for cind, cp in enumerate(cellpos):
        cells_raw[(tind,*cp)] = cell_bl[cind][tind] + cell_fluc[cind][tind]
        cells_bl[(tind,*cp)] = cell_bl[cind][tind]
    cells_raw[tind] = correlate(cells_raw[tind], morph.disk(rad))
    cells_bl[tind] = correlate(cells_bl[tind], morph.disk(rad))
    
shifts = [(-12,-12), (0,0) ,(12,12)]

cells_obs = cells_raw + bg
cells_bl += bg.mean()
cells_dff = (cells_obs - cells_bl) / (cells_bl)

for ind, s in enumerate(shifts):
    cells_obs[ind] = np.roll(cells_obs[ind], shift=s,axis=(0,1))
```

```python
fig, axs = plt.subplots(ncols=cells_raw.shape[0], nrows=3, dpi=200)
[ax.axis('off') for ax in axs.ravel()]
cmap = 'cividis'
for t in range(cells_raw.shape[0]):
    axs[0][t].imshow(cells_obs[t], cmap=cmap, clim=(cells_obs.min(), cells_obs.max()))    
    axs[-1][t].imshow((cells_dff)[t], cmap=cmap, clim=(cells_dff.min(), cells_dff.max()))    
```
